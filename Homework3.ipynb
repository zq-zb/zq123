4.用Python实现快速排序该算法，不允许直接调用插入排序函数，需要手动实现其逻辑过程。

```python
def quicksort(arr):
    if len(arr)<2:
        return arr
    else:
        x=arr[0]
        less=[i for i in arr[1:] if i<=x]
        mo=[i for i in arr[1:] if i>x]
        return quicksort(less)+[x]+quicksort(mo)
l=[9,5,4,3,2,1,8]
ar=quicksort(l)
print(ar)
```

5.使用 Python 随机生成三组长度递增的随机数列，长度分别为100、1000、10000，对生成的随机数组使用不同的排序算法，例如选择排序和归并排序，至少三种排序方法。对这些数列的数据排序，并分析不同排序算法在不同长度数列下的运行效果，例如运行时间等指标。

```python
import random
import time
def search(l):
    start_time=time.perf_counter();
    n=len(l)
    for i in range(n):
        minn=10001
        minx=0
        tmp=0
        for j in range(i,n):
          if minn>l[j]:
            minn=l[j]
            minx=j
        tmp=l[minx]
        l[minx]=l[i]
        l[i]=tmp
    end_time = time.perf_counter()  # 结束计时
    elapsed_time = end_time - start_time  # 计算运行时间
    print(f"运行时间: {elapsed_time:.6f} 秒")
    return l

def insert(l):
    start_time=time.perf_counter();
    n=len(l)
    for i in range(n):
      for j in range(i):
        if l[i-j]<l[i-j-1]:
            tmp=l[i-j]
            l[i-j]=l[i-j-1]
            l[i-j-1]=tmp
        else:
            break
    end_time = time.perf_counter()  # 结束计时
    elapsed_time = end_time - start_time  # 计算运行时间
    print(f"运行时间: {elapsed_time:.6f} 秒")
    return l

def quicksort(arr):
    if len(arr)<2:
        return arr
    else:
        x=arr[0]
        less=[i for i in arr[1:] if i<=x]
        mo=[i for i in arr[1:] if i>x]
        return quicksort(less)+[x]+quicksort(mo)

random1=[random.randint(1,1000) for _ in range(100)]
random2=[random.randint(1,10000) for _ in range(1000)]
random3=[random.randint(1,10000) for _ in range(10000)]

rand1=search(random1)
rand2=insert(random2)
start_time=time.perf_counter();
rand3=quicksort(random3)
end_time = time.perf_counter()  # 结束计时
elapsed_time = end_time - start_time  # 计算运行时间
print(f"运行时间: {elapsed_time:.6f} 秒")
print(rand1)
print(rand2)
print(rand3)
```

7.homework/data文件夹下的stuGrade.csv文件中包含5位同学的学号、语文成绩、数学成绩和英语成绩，请先用记事本打开该文件，查看文件内容，然后编写Python程序，读取该文件，并计算出各科的平均成绩（保留2位小数）。

```python
import csv

with open('stuGrade.csv',mode='r',newline='',encoding='utf-8') as file:
    csv_reader=csv.reader(file)
    next(csv_reader)
    sum1=0
    sum2=0
    sum3=0
    for row in csv_reader:
        sum1=sum1+(int)(row[1])
        sum2=sum2+(int)(row[2])
        sum3=sum3+(int)(row[3])
    ans1="{:.2f}".format(sum1*1.0/5)
    ans2="{:.2f}".format(sum2*1.0/5)
    ans3="{:.2f}".format(sum3*1.0/5)
    print(ans1,ans2,ans3)
```
